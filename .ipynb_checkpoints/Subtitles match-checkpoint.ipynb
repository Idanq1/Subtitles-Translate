```python
import pysrt
import pandas
from bs4 import BeautifulSoup
```

```python
heb_sub_path = r"Data\Hebrew\Mission.Impossible.Dead.Reckoning.Part.One.2023.1080p.BluRay.H264-RiSEHD.srt"
eng_sub_path = r"Data\English\Mission.Impossible.Dead.Reckoning.Part.One.2023.720p.BluRay.x264-ROEN.srt"
```

```python
heb_subs = pysrt.open(heb_sub_path, encoding="windows-1255")
eng_subs = pysrt.open(eng_sub_path)
```

```python
dif = 0.8
dif = int(dif * 1000)
```

```python
subs = pandas.DataFrame(columns=["Hebrew", "English", "Safe", "Hebrew_st", "Hebrew_e", "English_st", "English_e"])
for heb_sub in heb_subs:
    h_start = heb_sub.start
    h_end = heb_sub.end
    okay = True
    for eng_sub in eng_subs:
        e_start = eng_sub.start
        e_end = eng_sub.end

        if h_start + 2000 < e_start:
            break

        if h_start - dif > e_start:  # More than (dif) second apart
            continue
        if (e_end - dif) < h_end < (e_end + dif):
            if (e_start - dif) < h_start < (e_start + dif):
                heb_text = BeautifulSoup(heb_sub.text, "html.parser").getText()  # Removes html format like <i>
                eng_text = BeautifulSoup(eng_sub.text, "html.parser").getText()

                if heb_text.count("\n") != eng_text.count("\n"):  # Check if the numbers of lines is not the same - meaning chance of unmatching sentences
                okay = False
                break

```

```python
for sub in eng_subs:
    print(f"{sub.start} - {sub.end}")
    break
```
